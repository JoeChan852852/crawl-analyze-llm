# Objective

    We are presently poring over the map with meticulous care, endeavouring to verify 
    that every entry is both accurate and wholly unique—no small task, yet an essential one.

```python
import pandas as pd
Map = pd.read_csv('Locality_mapping.csv' ) # read the dataframe
Map
```

```python
Map['City'][202] = "Hong Kong"
Map['Continent'][236] = "Europe"
```

```python
Map.drop_duplicates(subset=['Country'])
```

    We have observed that certain entities appear under multiple guises within the dataset—most notably,
    “United States of America” alongside the abbreviated “United States”, and “United Kingdom of Great 
    Britain and Northern Ireland” juxtaposed with the more concise “United Kingdom”. Such heterogeneity 
    is, frankly, rather untidy. Accordingly, we must now undertake a thorough cleansing of the dataset 
    to enforce rigorous consistency and homogeneity across all such entries.

```python
Map['Country'] = Map['Country'].replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom')
Map['Country'] = Map['Country'].replace('England', 'United Kingdom')
Map['Country'] = Map['Country'].replace('United States of America', 'United States')
Map['Country'] = Map['Country'].replace('People’s Republic of China', 'China')
Map['Continent'] = Map['Continent'].replace('Australia', 'Oceania')
```

```python
Map.drop_duplicates(subset=['Country']).reset_index(drop=True)
```

```python
Map.drop_duplicates(subset=['Continent']).reset_index(drop=True)
```

    Now, having removed the duplicates, we can be confident that we no longer have alternative 
    spellings or variant names for the same country or continent.

```python
Map.to_csv('Locality_mapping_2025-11-21.csv')
```
