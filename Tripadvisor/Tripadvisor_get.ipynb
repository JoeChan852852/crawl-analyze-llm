```python
import pandas as pd
import WebCrawler
import os
import math
Topics = pd.read_csv('TA_2018-2019/Showforum 2018-2019/Topics_2018-2019.csv')
```

```python
from bs4 import BeautifulSoup
import re

def Check_replies(html):
    # Parse the HTML
    soup = BeautifulSoup(html, 'html.parser')
    # Find the div with class "pgCount"
    pg_count_div = soup.find('div', class_='pgCount')

    if pg_count_div:
        # Get the text content of the div
        text = pg_count_div.get_text(strip=True)  # "1-10of16 replies"
        replies = text.split("of")
        replies = int(re.sub(r'\D', '',replies[1]))  
    return replies
```

```python
lower_bound = 1
upper_bound = 5
while lower_bound < upper_bound:
    folder_path = str(lower_bound)
    os.makedirs(folder_path, exist_ok=True)
    target = f"https://www.tripadvisor.com{Topics["Topics"][lower_bound]}"
    print(f"Downloading <<<<{target}")
    HTML_name =  folder_path + Topics["Topics"][lower_bound]
    print(HTML_name)
    source = WebCrawler.HTML_get(target,HTML_name)
    number_of_replies = Check_replies(source)
    print(f"Number of replies >>>> {number_of_replies}")
    subpage_no = math.ceil(number_of_replies/10)
    print(f"total number of page >>>> {subpage_no}")
    if subpage_no > 1:
        sub_link = target.split("-")
        sub_link = sub_link[0]+ "-" + sub_link[1] + "-" + sub_link[2] + "-" + sub_link[3]
        for i in range(subpage_no):
            if i > 0:
                target = sub_link + "-o" + str(i*10)
                print(f"Downloading sub link <<<< {target}")
                HTML_name = target.split(".com")
                HTML_name = folder_path + HTML_name[1]
                WebCrawler.HTML_get(target,HTML_name)
    lower_bound = lower_bound + 1
    
```
