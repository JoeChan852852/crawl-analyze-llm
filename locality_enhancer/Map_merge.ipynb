```python
import pandas as pd
Map = pd.read_csv('Locality_mapping_2025-12-20.csv') # read the dataframe
TA = pd.read_csv('TAPosts_2023.csv')
```

```python
TA
```

```python
# Extract column 'B' into a new DataFrame df_B
locality = TA[['locality']].copy()
TA.drop
locality
```

```python
Map
```

```python
merged_df = pd.merge(locality, Map, on='locality',how='left')
merged_df
```

```python
# Define the insertion point (index 1)
insert_loc = 4

# Split df1 into two parts
df1_part1 = TA.iloc[:, :insert_loc]  # Columns from start to loc-1
df1_part2 = TA.iloc[:, insert_loc:]  # Columns from loc to end

# Concatenate the parts in the desired order
# Axis=1 ensures column-wise concatenation
df_final = pd.concat([df1_part1, merged_df, df1_part2], axis=1)

print(df_final)
```

```python
df_final.to_csv('TA2023_Location.csv', index=False)
```
