# Introduction

    This programme is designed to retrieve data from Instagram, either from a specific 
    user profile or from posts associated with a particular hashtag.

# Set up

    We must first log in to Instagram via a web browser (Firefox!) in order to obtain an authenticated 
    session cookie. This cookie will subsequently allow us to mimic a genuine logged-in user 
    when making requests, thereby bypassing certain restrictions imposed on unauthenticated 
    access.

```python
'''
To create a login session. Login to the ac on Firefox before running this cell
'''
import instaloader
import time
import random
import browser_cookie3
import pandas as pd
import csv
import os

# Create an instance of Instaloader
loader = instaloader.Instaloader()

# Login to an IG ac

loader.login("trashig001", "aA30624700")
# To emulate as a real user we need to access the cookie  on the browser
# Login to the the IG ac on the browser to make sure we get the cookie 
loader.context._session.cookies = browser_cookie3.firefox(domain_name='instagram.com')
```

    To illustrate the process, let us assume we wish to extract data from the Instagram profile 
    "visitjapanjp". The initial step consists of retrieving the shortcodes of all posts belonging 
    to that profile and storing them in a list. These shortcodes will subsequently serve as 
    unique identifiers, enabling us to fetch detailed information and media for each individual
    post at a later stage.

```python
'''
Replace with the target Instagram username in a list. These shortcodes will subsequently serve 
as unique identifiers, us to fetch detailed information and media for each individual post at 
a later stage.
'''
target_profile = "visitjapanjp"  # Specify the target Instagram username (profile) you want to scrape posts from.

# Load the target profile
profile = instaloader.Profile.from_username(loader.context, target_profile)

# lower bound of posts
m = 0 
# upper bound of posts 
n = 10 # Assume we take 10 posts 
# Initialize a list to store post shortcodes.
postcodes = []

print(f"Fetching posts for profile: {target_profile}")

# Retrieve posts from the profile and collect the shortcode for each post.
for post in profile.get_posts():
    postcodes.append(post.shortcode)
    print(f"Found post: https://www.instagram.com/p/{post.shortcode}/")
    m = m + 1
    if m == n :
        break
print(f"\nTotal posts found: {len(postcodes)}")
```

## Initialize

```python
# Initialize empty DataFrame with column names
df = pd.DataFrame(columns=['postcode', 'Date', 'Number of likes', 'Number of comments', 
                           'Owner Username','Media Count','Photo Count' , 'Video Count','Content'])

'''
For Media Counting
'''
# Common file extensions for photos and videos
photo_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'}
video_extensions = {'.mp4', '.avi', '.mov', '.wmv', '.flv', '.mkv'}
```

## Data Get

```python
import random
# starting point
i = 0
# download the pictures comment and post info to a folder
while i < (len(postcodes)):
    random_integer = random.randint(1, 3)
    n = str(i)
    shortcode = postcodes[i]
    post = instaloader.Post.from_shortcode(loader.context, shortcode)
    loader.download_post(post, target=n)
    file = n + "/" + n + ".txt"
    f = open(file, "x")
    print(shortcode, file=f)
    print("Post Details:", file=f)
    print("Caption:\n", post.caption , file=f)
    print("Date:", post.date , file=f)
    print("Number of likes:", post.likes , file=f)
    print("Number of comments:", post.comments, file=f)
    print("Owner Username:", post.owner_username, file=f)
    print("-"*40, file=f)
    print("Comments:", file=f)
    photo_count = 0
    video_count = 0
    
    for media in os.listdir(n):
        # Get the file extension (converted to lowercase)
        file_ext = os.path.splitext(media)[1].lower()

        # Count photos
        if file_ext in photo_extensions:
            photo_count += 1
        # Count videos
        elif file_ext in video_extensions:
            video_count += 1
    
    row = {
        'postcode': shortcode,
        'Date': post.date,
        'Number of likes': post.likes,
        'Number of comments': post.comments,
        'Owner Username': post.owner_username,
        'Media Count' : post.mediacount,
        'Photo Count' : photo_count,
        'Video Count' : video_count,
        'Content' : post.caption
    }
    df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
    
    for comment in post.get_comments():
            username = comment.owner.username   # Username of the commenter
            text = comment.text                  # The comment text
            print(f"User: {username} \nComment: {text}\n{'-'*40}" ,file=f)
    f.close()
    time.sleep(random_integer)
    i = i + 1 
```
